QNA for MassHealth (Standard)

FRAMING:

For my final project, I created a QNA to help potential users determine if they are likely eligible for MassHealth (Standard). In order to help users find eligibility, I had to navigate the QNA around Residency, Income, Household, Citizenship, and other requirements. At one point, the QNA became complicated enough that the Google Chrome Web Browser could no longer handle to set of code. 

This QNA is specifically meant for Law Practitioners to use with clients, but I have organized it in such a way that a client could go through the QNA on their own if they felt confident to do so. This allows for multiple user-sets so that Lawyers (and maybe social workers) can help their clients determine eligibility in an organized and efficient way. I debated whether to link the answers to a form, in order to spit out the relavant information at the end of the QNA when the lawyer and client are ready to actually apply for MassHealth (Standard) if they are more likely to be eligible than not.

In general, signing up for MassHealth is not easy. Both practitioners and their clients find the process overwhelming because there are so many different requirements and so many different types of coverage. Even within one type of coverage, you have residency, income, household, citizenship, and qualifying-person requirements that are very specific and hard to keep track of. Also, practitioners don't always need to help clients with signing up for benefits, so when they do partake in the sign-up process, they are often as unfamiliar with the Massachusetts benefits options as their clients. 

As a law student and future practitioner, I was also very unaware of the MassHealth requirements. We did not learn about benefits in any of our core classes and I am unaware of a supplemental class which focuses on benefits and related access-to-justice issues; I was only familiar with MassHealth on a word-of-mouth basis. In the Family Advocacy Clinic, we work with low-income clients, most of whom need Health Insurance and whose quality of life would improve if they were signed up for different types of benefits and different forms of MassHealth coverage. Fortunately for users (and unfortunately for QNA purposes), the coverage options offered by MassHealth are so extensive that it would be too overwhelming to create an QNA for the entirety of the MassHealth system in one semester. However, by creating a QNA for MassHealth (Standard) for this semester, a coverage-type which typically has no cost for users, low-income clients are best served and practitioners have a tool to help clients in need who cannot afford to pay costly hourly fees to the lawyer. This QNA streamlines the process, allowing the lawyer to dedicate a smaller amount of time to the client (in order to avoid heavy billing) but with the same if not greater output. 

The relavant stakeholders/users are legal practitioners with low-income clients who need health insurance or who need lower-cost health insurance. The clients themselves are also users, in theory, but MassHealth (Standard) is so complicated and so particular about who can receive coverage that it is better for a client to work with a lawyer or someone who has minimal legal expertise. For clients who choose to participate in the QNA without a lawyer and for lawyers who may need a little more assistance, I have included phone numbers and web links within the QNA. 

Other potential users include anyone working with low-income clients, including social workers, therapists, and anyone else who serves low-income communities.

IDEATION & PROTOTYPING:

Of the projects we created in Coding the Law this semester, I found the QNA style of code to be the most user friendly for determining Benefits eligibility. Especially for MassHealth, a tool which asked a lot of clarifying questions seemed to be the best device to gather specific client information efficiently and effectively. The Data-Sets Project did not seem like a good fit because the relavent data is not needed to help clients sign up for MassHealth, only the eligibility requirements. Also, the TwitterBot project would not have been useful because MassHealth information and requirements do not change that frequently to require something that regularly updates a user.

To test that this QNA idea (to determine Benefits eligibility) would be beneficial, and as effective as predicted, I created a decision tree of the various requirements to map out the QNA's trajectory. The decision tree, though filled with various, twisting branches, was organized enough on a first draft that I felt confident about moving forward with the QNA. The QNA would be able to cover all of the relevant requirements and could account for moments where more information was needed, mainly by providing phone numbers and website for practitioners and clients to use in order to clarify any confusion. MassHealth (Standard) is tricky because, even if you seem eligible for that type of coverage, you may not be. Moments of confusion during the QNA are either easy enough to clear up with an explanation or the individual would need to contact a MassHealth Representative anyway for clarification, with or without using the help of legal practitioner. 

I am not aware of any other types of models which would otherwise best suit the situation. As I mentioned, I reviewed all of my other options before determining that I wanted to create a QNA and the QNA remained the best device to accomplish my goal: giving people a tool to help clients apply for MassHealth (Standard) Coverage. 


USER TESTING/FEEDBACK:

All feedback for the project occurred in person. I did not think to incorporate a feedback form until the last minute and found the in-person feedback more helpful than creating a form.

In regards to feedback, I incorporated a lot of changes to questions based on (1) user confusion, (2) suggested clarifications, (3) and general bugs with the QNA. I received feedback from multiple different types of users. One of my first users was a potential real-world user, medium income woman in her early/mid 20s with no law experience. This woman is new to Massachusetts and is looking to get off of her parents health insurance. She was wondering if she qualified for MassHealth Standard because she does not make a lot of money, her parents pay more than they can afford for her health insurance, and she is taking a break from school so she is worried about being kicked off of her parents health insurance. When we went through the QNA, she had a lot of questions about what "Residency" required and what a modified adjusted gross income was. I incorporated clarifications for her and users like her. We also encountered a bug with the math, which I was able to correct/fix with the help/technological expertise of Professor Colarusso. We determined, in the end, that she was not eligible for MassHealth (Standard). She is over-income for MassHealth (Standard). However, because I had links to the eligibility packet, I was able to quickly look up which coverage she would be eligible for. She was both satisfied by the QNA's help and by my ability to help her further due to the links I provided in the QNA itself.

I also received feedback from several law students/future lawyers due to the fact that they work with low-income clients and have an idea of what information they would need to best help those clients. There were times where the law students wanted some clarifying pieces of information in the QNA so that they could describe to their clients exactly what type of information they were looking for. The feedback from law students was relatively positive. Several students even asked me to send them the QNA questions and link in the future (once the project is more completed) so that they can run through the QNA with their clients in practice. 

I do not have a copy of user feedback, as all of the user feedback was oral. I have typed all of the feedback I was given in this notebook. If you require a specific feedback form, I am happy to create one and have the previous users fill the form out. I chose to spend the time editing the QNA instead so that I could incorporate feedback quickly. Every component of the QNA was edited (sometimes substantially) at least once thanks to the feedback I received from the potential real-world user, the 20+ year old woman, running the QNA with the information of two of my clients, and the user feedback from other law students. In the coming semester, I intend to send the more finalized QNA to higher-stakes users, such as Clinic Supervisors, Pro-Bono Attorneys, and other practitioners who work with low and medium income clients. 

REFINEMENT:

I integrated ALL feedback from user testing because I was the QNA questions to be as clear and concise as possible. I don't want users to get confused or overwhelmed because that would lead to demotivation, which would negatively impact the work-product and efficiency of the user while running the QNA which could in turn affect the client's outlook on the situation. 

Every section received major changes. Certain sections were broken down into smaller sections so that the material provided was less confusing and overwhelming, such as the Citizenship section. The Residency section originally did not have as many definitons, clarifications, or options, but feedback indicated that all three of those additions were required. I conducted multiple rounds of user testing, with several law students (at least four law students), one potential real-world user (the 20+ year old woman), and using the scenarios that match my clients in the Family Advocacy Clinic in order to test of efficient and effective the QNA is. Due to my testing the information provided for me by my clients, I was able to clarify and incorporate different points regarding the Citizenship section of the QNA. The multiple tests were done specifically to improve the QNA and the fix any bugs that I may have missed, not simply for the sake of testing because that would be a waste of time.

I intend to continue to test the MassHealth QNA in the coming semester and the incorporate even more feedback in order to further test and further finalize the QNA for real-world use. The QNA is currently ready for real-world use, as is, but it would be even more efficient and effective with further user testing. I want it to be the best possible product.
